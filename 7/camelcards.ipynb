```python
file_path = r'C:\Users\KG\Documents\GitHub\advent-of-code-2023\7\hands.txt' 

with open(file_path, 'r') as file:
    orgiginal_lines = file.readlines()

orgiginal_lines = [line.strip() for line in orgiginal_lines]
orgiginal_lines
```

```python
hands = [tuple(l.split(" ")) for l in orgiginal_lines]
hands
```

```python

def get_rank(hand):
    count = {}
    for c in hand:
        count[c] = count.get(c, 0) + 1
    scores = list(reversed(sorted(list(count.values()))))
    if scores[0] == 5:
        return 6
    if scores[0] == 4:
        return 5
    if scores[0] == 3 and scores[1] == 2:
        return 4
    if scores[0] == 3:
        return 3
    if scores[0] == 2 and scores[1] == 2:
        return 2
    if scores[0] == 2:
        return 1
    return 0

```

```python
ranked_hands = {}
for hand, bid in hands:
    rank = get_rank(hand)
    rank_hands = ranked_hands.get(rank, [])
    rank_hands.append((hand, bid))
    ranked_hands[rank] = rank_hands
ranked_hands
```

```python
def in_rank_rank(rank_hands):
    return int( rank_hands.replace("A", "E").replace("T", "A").replace("J", "B").replace("K", "C").replace("Q", "D"), 30) 
in_rank_rank('TT')
```

```python
for rank, hands in ranked_hands.items():
    ranked_hands[rank] = sorted(hands, key = lambda hand: in_rank_rank(hand[0]))
ranked_hands
```

```python
hands_listed = []
for rank in sorted(ranked_hands.keys()):
    print(ranked_hands[rank][0], ranked_hands[rank][-1])
    hands_listed.extend(ranked_hands[rank])

    

```
