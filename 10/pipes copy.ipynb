```python
def fuck_you(c):
    if c == "-":
        return set(["l", "r"])
    if c == "|":
        return set(["u", "d"])
    if c == "L":
        return set(["u", "r"])
    if c == "J":
        return set(["u", "l"])
    if c == "F":
        return set(["r", "d"])
    if c == "7":
        return set(["l", "d"])
    return set()
```

```python


class CustomList:
    def __init__(self, start):
        self.items = start

    def __getitem__(self, index):
        if isinstance(index, slice):
            # If a slice is provided, return a new CustomList with sliced elements
            return CustomList(self.items[index])
        elif 0 <= index < len(self.items):
            return self.items[index]
        else:
            return None



    def __repr__(self):
        return repr(self.items)
```

```python
def oposite(d):
    if d == "r":
        return "l"
    elif d == "l":
        return "r"
    elif d == "d":
        return "u"
    elif d == "u":
        return "d"
def map_pipe(start, map):
    map = CustomList([CustomList(l) for l in map])
    pipes = [start]
    direction = "r"
    x = start[1]
    y = start[0]
    while True:
        if direction == "r":
            x += 1
        elif direction == "l":
            x -= 1
        elif direction == "d":
            y += 1
        elif direction == "u":
            y -= 1
        if (y,x) == start:
            break
        pipes.append((y,x))
        direction = list(map[y][x] - {oposite(direction)})[0]
    return pipes


        

pipes = map_pipe(s_pos, set_map)
pipes
```

```python
def fill(map, start):
    
    #map = CustomList([CustomList(l[:]) for l in map])
    frontier = [start]
    while frontier != []:
        new_frontier = []
        for f in frontier:
            x = f[1]
            y = f[0]
            search = [(y-1, x), (y, x+1), (y+1, x), (y, x-1)]
            for s in search:
                if len(map) > s[0] and len(map[s[0]]) > s[1] and map[s[0]][s[1]] == ".":
                    map[s[0]][s[1]] = "I"
                    new_frontier.append(s)
        frontier = new_frontier
    return [l[:] for l in map]
```
