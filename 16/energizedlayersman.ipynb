```python
file_path = r'C:\Users\KG\Documents\GitHub\advent-of-code-2023\16\mirrormap.txt' 

with open(file_path, 'r') as file:
    orgiginal_lines = file.readlines()


orgiginal_lines = [line.strip() for line in orgiginal_lines]
for l in orgiginal_lines:
    print(l)
```

```python
import numpy as np
mirror_map = np.array([np.array(list(l)) for l in orgiginal_lines])
mirror_map
```

```python
import math
def cos(d):
    return int(round(math.cos(math.radians(d)),1))
def sin(d):
    return int(round(math.sin(math.radians(d)),1))
cos(180)
```

```python
energized_map = mirror_map.copy()
memmory_map = np.full(tuple(list(mirror_map.shape) + [4]), np.array([False,False,False,False]))
import math
rays = [(0,-1,90)]
energized_map[0,0] = 90
while len(rays) != 0:
    new_rays = []
    for y,x,d in rays:
        if memmory_map[y,x,d//90]:
            continue
        memmory_map[y,x,d//90] = True
        energized_map[y,x] = mirror_map[y,x]
        if d == 0:
            y -= 1
        elif d == 90:
            x += 1
        elif d == 180:
            y += 1
        elif d == 270:
            x -= 1
        if y < 0 or y >= mirror_map.shape[0] or x < 0 or x >= mirror_map.shape[1]:
            continue
        energized_map[y,x] = d
        if mirror_map[y,x] == "|":
            if d == 90 or d == 270:
                new_rays.append((y,x, 0))
                new_rays.append((y,x, 180))
            else:
                new_rays.append((y,x,d))
        elif mirror_map[y,x] == "-":
            if d == 0 or d == 180:
                new_rays.append((y,x, 90))
                new_rays.append((y,x, 270))
            else:
                new_rays.append((y,x,d))
        elif mirror_map[y,x] == "/":
            if d == 0:
                new_rays.append((y,x,90))
            elif d == 90:
                new_rays.append((y,x,0))
            elif d == 180:
                new_rays.append((y,x,270))
            elif d == 270:
                new_rays.append((y,x,180))
        elif mirror_map[y,x] == "\\":
            if d == 0:
                new_rays.append((y,x,270))
            elif d == 90:
                new_rays.append((y,x,180))
            elif d == 180:
                new_rays.append((y,x,90))
            elif d == 270:
                new_rays.append((y,x,0))
        else:
             new_rays.append((y,x,d))
    rays = new_rays
    print((energized_map))

np.sum(np.any(memmory_map, 2))
```
