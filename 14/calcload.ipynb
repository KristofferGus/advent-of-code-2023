```python
import functools

def memoize(func):
    cache = {}

    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        key = (args, frozenset(kwargs.items()))
        if key not in cache:
            result = func(*args, **kwargs)
            cache[key] = result
        return cache[key]

    return wrapper
```

```python

@memoize
def go_up(column):
    column = np.array(column)
    steady_rocks = np.where(column == "#")
    steady_rocks = np.insert(steady_rocks, 0, -1)
    steady_rocks = np.insert(steady_rocks, steady_rocks.shape[0], column.shape[0])
    new_column = np.array([])
    for start, end in zip(steady_rocks[:-1], steady_rocks[1:]):
        section = column[start+1:end]
        rolled_rocks = np.sum(section == "O")
        empty = end - (start + 1) - rolled_rocks
        new_column = np.concatenate((new_column, np.full(rolled_rocks, "O"), np.full(empty, "."), np.array(["#"])))
    return new_column[:-1]

go_up(tuple(rocks[:,0]))

print(rocks)
rolled_rocks = np.stack([go_up(tuple(c)) for c in rocks.T]).T
print(rolled_rocks)
```

```python
suum = 0
for i, row in enumerate(rolled_rocks):
    suum += (rolled_rocks.shape[0] - i) * np.sum(row == "O")
suum
```
