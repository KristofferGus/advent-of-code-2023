```python
import numpy as np

space = 1000000

lines = np.array([list(l[:]) for l in orgiginal_lines])
row_dist = np.ones(lines.shape[0], dtype=np.int64)
colum_dist = np.ones(lines.shape[1], dtype=np.int64)
print(lines.shape)
index = 0
while index < lines.shape[0]:
    row = lines[index]
    if all([row[0] == i for i in row]):
        row_dist[index] = space
        index += 1
    index += 1

index = 0
while index < len(lines[0]):
    column = lines[:,index]
    if all([column[0] == i for i in column]):
        colum_dist[index] = space
        index += 1
    index += 1

      

print(row_dist, colum_dist)  
```

```python
sum(row_dist[0:100])
```

```python
galaxies = [(x, y) for x, row in enumerate(lines) for y, element in enumerate(row) if element == "#"]
galaxies
```

```python
pairs = set()
for a in galaxies:
    for b in galaxies:
        if not (a == b):
            pairs.add(tuple(sorted([a,b])))
        


pairs
```

```python
def calc_cool_distance(a, b):
    x1, x2 = sorted([a[1], b[1]])
    y1, y2 = sorted([a[0], b[0]])
    return np.sum(row_dist[y1:y2]) + np.sum(colum_dist[x1:x2])

print(calc_cool_distance((0, 3), (9, 0)))
print(calc_cool_distance((9, 0), (0, 3)))
print(calc_cool_distance((0, 3), (4, 6)))
```

```python
all([a == b for a,b in zip([calc_cool_distance(b,a) for a,b in pairs], [calc_cool_distance(a,b) for a,b in pairs])])
```
