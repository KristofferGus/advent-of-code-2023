```python
digits = ['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
digits = list(zip(digits, range(1, len(digits) + 1)))
digits
```

```python
teststring = "twonefive2threekfrtvnbmplpsevenseven"
def replace_digits(strang: str):
    out = ""
    i = 0
    while(i < len(strang)):
        was_digit = False
        for digit in digits:
            if strang[i:i + len(digit[0])] == digit[0]:
                i += len(digit[0])
                out += str(digit[1])
                was_digit = True
                break
        if not was_digit:
            out += strang[i]
            i +=1
    return out

  
test_fixed = replace_digits(teststring)
test_fixed
```

```python
fixed_lines = [replace_digits(line) for line in lines]
[t for t in list(zip(fixed_lines, lines)) if t[0] == t[1]]
list(zip(fixed_lines, lines))
```

```python
pairs = []
for line in fixed_lines:
    first = ""
    last = ""
    for c in line:
        if c.isdigit():
            if first == "":
                first = c
            last = c
    pairs.append((first,last))
list(zip(pairs, lines))
```

```python
calnumbs = [int(p[0] + p[1]) for p in pairs]
calnumbsum = sum(calnumbs)
calnumbsum
```

```python
pairs = []
for line in lines:

    substrings = []
    for i in range(len(line) + 1):

        for j in range(i,len(line) + 1):
            substrings.append(line[i:j])
    first = ""
    last = ""
    for sub in substrings:
        for digit in digits:
            if(sub == digit[0]):
                if(first == ""):
                    first = digit[1]
                last = digit[1]
            
            if(sub == str(digit[1])):
                if(first == ""):
                    first = digit[1]
                last = digit[1]
    pairs.append((first, last))
list(zip(pairs, lines))  
```

```python
numbs = [int(str(p[0]) + str(p[1])) for p in pairs]
sum(numbs)
```
