```python
lines = orgiginal_lines[0].split(",")
lines
```

```python
def get_ascii_values(input_string):
    ascii_values = [ord(char) for char in input_string]
    return ascii_values
get_ascii_values("HASH")
```

```python
def calc(numbs):
    currenct_value = 0
    for num in numbs:
        currenct_value += num
        currenct_value *= 17
        currenct_value = currenct_value % 256
    return currenct_value
calc(get_ascii_values("HASH"))
```

```python
sum([calc(get_ascii_values(l)) for l in lines])
```

```python
def make_instuction(line):
    if line[-1] == "-":
        return ("-", calc(get_ascii_values(line[:-1])), line[:-1])
    word, strength = tuple(line.split("="))
    return ("=", calc(get_ascii_values(word)),word, strength )
make_instuction(lines[0])
make_instuction(lines[1])
```

```python
instructions = [make_instuction(l) for l in lines]
instructions
```

```python
boxes = [[] for i in range(256)]
for instruction in instructions:
    if instruction[0] == "=":
        box_number, label, strength = tuple(list(instruction)[1:])
        lenses = boxes[box_number]

        found = False
        for index, (lens_label, _) in enumerate(lenses):
            if lens_label == label:
                lenses[index] = (label, strength)
                found = True
                break

        if not found:
            lenses.append((label,strength))
        boxes[box_number] = lenses
    else:
        box_number, label = tuple(list(instruction)[1:])
        lenses = boxes[box_number]
        new_lenses = []
        for lesn_label, strength in lenses:
            if not label == lesn_label:
                new_lenses.append((lesn_label, strength))
        boxes[box_number] = new_lenses
boxes        
```

```python
suum = 0
for box_numb, lenses in enumerate(boxes):
    for lens_index,( _, strength) in enumerate(lenses):
        suum += (box_numb + 1) * (lens_index + 1) * int(strength)
suum
```
